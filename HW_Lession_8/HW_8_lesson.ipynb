- Воспользуемся правилом нахождения производной от частного: 


$$\frac{d}{d x}\left(\frac{u}{v}\right)=\frac{v \frac{d u}{d x}-u \frac{d v}{d x}}{v^{2}}$$
Где $u = x$ и $ v = |x| + 1$.

Следовательно, в нашем случае имеем:

$\frac{d}{d x}\left(\frac{x}{1+|x|}\right)= \frac{(1+|x|)\left(\frac{d}{d x}(x)\right)-x\left(\frac{d}{d x}(1+|x|)\right)}{(1+|x|)^{2}}= \frac{1+|x|-x\left(\frac{d}{d x}(1+|x|)\right)}{(1+|x|)^{2}}=\frac{1+|x|-x\left(\frac{d}{d x}(|x|)+0\right)}{(1+|x|)^{2}} = $

- Запишем $\frac{d}{d x}(|x|)$ след. образом  $\frac{d}{d x}(|x|)=\frac{d|u|}{d u} \frac{d u}{d x}$, где $u=x$, $\frac{d}{d u}(|u|)=\frac{u}{|u|}$

$= \frac{1+|x|-\frac{x\left(\frac{d}{d x}(x)\right)}{|x|}}{(1+|x|)^{2}}= \frac{1+|x|-\frac{x^{2}\left(\frac{d}{d x}(x)\right)}{|x|}}{(1+|x|)^{2}}= \frac{1+|x|-1 \mid \frac{x^{2}}{|x|}}{(1+|x|)^{2}}= \frac{1}{(1+|x|)^{2}}$

$$f'(x) = \frac{1}{(1+|x|)^{2}}$$

3. **Hyperbolic tangent/Гиперболический тангенс**. $f(x)=\operatorname{th}(x)=\frac{\left(e^{x}-e^{-x}\right)}{\left(e^{x}+e^{-x}\right)}$
4. **ReLU**. $f(x)=\left\{\begin{array}{ll}0 & x<0 \\ x & x \geqslant 0\end{array}\right.$

1. **Identity.** $f(x)=x$

Производная:

$$ f'(x)=x'=1 $$

Область определения:
$$ D(f) = \mathbb{R} $$

Функция монотонно возрастающая

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x

a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Identity')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

2. **Binary step/Единичная ступенька**. $f(x)=\left\{\begin{array}{ll}0 & x<0 \\ 1 & x \geqslant 0\end{array}\right.$

Производная $$ f'(x) = 0, при  x 	$\neq$0 $$

Область определения:
$$ D(f) = \mathbb{R} $$

Функция кусочно-постоянная



```
def f(x):
   return [0 if i < 0 else 1 for i in x]

a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('Binary step')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

3. **Hyperbolic tangent/Гиперболический тангенс**. $f(x)=\operatorname{th}(x)=\frac{\left(e^{x}-e^{-x}\right)}{\left(e^{x}+e^{-x}\right)}$


Производная: $$ f'(x) = sech^2x $$
Область определения:
$$ D(f) = \mathbb{R} $$

Монотонно возрастающая

```
def f(x):
   return np.tanh(x)

a = np.arange(-20, 20, 0.1)
plt.plot(a, f(a))
plt.title('Hyperbolic tangent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

4. **ReLU**. $f(x)=\left\{\begin{array}{ll}0 & x<0 \\ x & x \geqslant 0\end{array}\right.$

Производная $f'(x)=\left\{\begin{array}{ll}0 & x<0 \\ 1 & x \geqslant 0\end{array}\right.$

Область определения:
$$ D(f) = \mathbb{R} $$

при x< 0 функция постоянная, при x >=0 функция монотонно возрастает

```
def f(x):
   return [0 if i < 0 else i for i in x]

a = np.arange(-100, 100)
plt.plot(a, f(a))
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

5. **Softplus**. $f(x)=\ln \left(1+e^{x}\right)$

Производная: $$ f'(x) = \frac{e^x}{1+e^x}$$
Область определения:
$$ D(f) = \mathbb{R} $$

Функция монотонно возрастает

```
def f(x):
   return np.log(1+np.e**x)

a = np.arange(-10, 10, 0.01)
plt.plot(a, f(a))
plt.title('Softplus')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

6. **Синусоида**. $f(x)=\sin (x)$

Производная: $$ f'(x) = cosx $$
Область определения:
$$ D(f) = \mathbb{R} $$

Функция монотонно возрастает на отрезках $$ [ - \pi/2 + 2\pi k; \pi/2+2\pi k]$$ и убывает на $$ [ \pi/2 + 2\pi k; 3\pi/2+2\pi k] ,k \in \mathbb{Z} $$ 

```
def f(x):
   return np.sin(x)

a = np.arange(-10, 10, 0.1)
plt.plot(a, f(a))
plt.title('Sinus')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
